```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score, recall_score, precision_score, average_precision_score, f1_score, classification_report, accuracy_score, plot_roc_curve, plot_precision_recall_curve, plot_confusion_matrix
```

```python
df = pd.read_csv('londonweekends.csv', na_values='?')
#df.head()
df.columns
```

```python
df = df.rename(columns={'realSum': 'target'})
df['target'].value_counts(dropna=False)
```

```python
df = pd.get_dummies(df, columns=['biz', 'multi'], drop_first=True)
df
```

```python
numeric_cols = ['person_capacity',  'cleanliness_rating',
       'guest_satisfaction_overall', 'bedrooms', 'dist', 'metro_dist',
       'attr_index', 'attr_index_norm', 'rest_index', 'rest_index_norm', 'lng',
       'lat']
cat_cols = list(set(df.columns) - set(numeric_cols) - {'target'}) #categorical columns
cat_cols.sort()

print(numeric_cols)
print(cat_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  
sns.set_theme(style="ticks", rc={"axes.spines.right": False, "axes.spines.top": False})
 
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['dist'], y=df['target'], color="red",alpha=0.2)
plt.ylabel('prices')
plt.title("Capacity vs price")
plt.ylim(0, 2000)
 
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['metro_dist'], y=df['target'], color="green",alpha=0.4)
plt.ylabel('prices')
plt.title("rating vs Price")

plt.ylim(0, 2000)
plt.show()
```
